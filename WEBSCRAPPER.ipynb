```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

req=requests.get("http://www.useams.org/")
soup=BeautifulSoup(req.content,"html.parser")
print(soup.prettify())
f=soup.find("div", attrs={'class':'adr'}).get_text()
html=str(f)

with open("myfile1.txt", "w") as file1:
    # Writing data to a file
    file1.write(html)
    #file1.writelines(L)
#with pd.ExcelWriter("zzz.xlsx", mode="w+",engine="openpyxl") as writer:
#    df.to_excel(writer,sheet_name="Sheet1")

```
